# **Вводные**

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import shapiro

df = pd.read_excel("data5.xlsx", sheet_name=1, index_col='state_name', na_values='..')

# оставляем страны, по которым есть полные данные с 2003 по 2022
df = df.groupby('state_name').filter(lambda x: x['year'].min() <= 2002 and x['year'].max() >= 2022)

# удаляем чрезмерно экономически развитые страны (выбросы)
df = df.drop(['Bahrain', 'Kuwait', 'Oman', 'Qatar', 'Saudi Arabia', 'Trinidad and Tobago', 
              'United Arab Emirates', 'Chile', 'Croatia', 'Hungary', 'Poland', 'Uruguay', 'Equatorial Guinea'])

# удаляем Океанию
df = df[df['region'] != 7]

# выносим названия некоторых колонок в отдельный список для последующих манипуляций
cols = ['libdem_norm_diff', 'property_rights_diff', 'tax_burden_diff', 'government_spending_diff', 
        'business_freedom_diff', 'monetary_freedom_diff', 'trade_freedom_diff', 'investment_freedom_diff', 
        'financial_freedom_diff', 'gov_effectiveness_diff', 'corruption_diff', 'pstab_diff']

cols_1 = ['libdem_norm', 'property_rights', 'tax_burden', 'government_spending', 
        'business_freedom', 'monetary_freedom', 'trade_freedom', 'investment_freedom', 
        'financial_freedom', 'gov_effectiveness', 'corruption', 'pstab']

# делаем отдельный df для оценки распределений переменных-разниц
df1 = df[df['year'] != 2002].copy()
```

# **Оценки распределения по группам**

```python
df_11_plus = df[df.groupby('state_name')['imf_pr'].transform('sum') >= 11]
df_5_to_10 = df[(df.groupby('state_name')['imf_pr'].transform('sum') >= 5) & (df.groupby('state_name')['imf_pr'].transform('sum') <= 10)]
df_4_minus = df[df.groupby('state_name')['imf_pr'].transform('sum') <= 4]

df_11_plus_1 = df1[df1.groupby('state_name')['imf_pr'].transform('sum') >= 11]
df_5_to_10_1 = df1[(df1.groupby('state_name')['imf_pr'].transform('sum') >= 5) & (df1.groupby('state_name')['imf_pr'].transform('sum') <= 10)]
df_4_minus_1 = df1[df1.groupby('state_name')['imf_pr'].transform('sum') <= 4]
```

```python
for col in cols_1:
    print('11+ лет:')
    print(df_11_plus[col].describe())
    print()
    print('От 5 до 10 лет:')
    print(df_5_to_10[col].describe())
    print()
    print('Меньше 4 лет:')
    print(df_4_minus[col].describe())
    print()
    print('11+ лет:')
    print()
    stat, p = shapiro(df_11_plus[col])
    print(f"Shapiro-Wilk test: statistic={stat:.4f}, p-value={p:.4f}")
    sns.histplot(df_11_plus[col])
    plt.show()
    print()
    print('От 5 до 10 лет:')
    print()
    stat, p = shapiro(df_5_to_10[col])
    print(f"Shapiro-Wilk test: statistic={stat:.4f}, p-value={p:.4f}")
    sns.histplot(df_5_to_10[col])
    plt.show()
    print('Меньше 4 лет:')
    print()
    stat, p = shapiro(df_4_minus[col])
    print(f"Shapiro-Wilk test: statistic={stat:.4f}, p-value={p:.4f}")
    sns.histplot(df_4_minus[col])
    plt.show()
```

```python
for col in cols:
    print('11+ лет:')
    print(df_11_plus_1[col].describe())
    print()
    print('От 5 до 10 лет:')
    print(df_5_to_10_1[col].describe())
    print()
    print('Меньше 4 лет:')
    print(df_4_minus_1[col].describe())
    print()
    print('11+ лет:')
    print()
    stat, p = shapiro(df_11_plus_1)
    print(f"Shapiro-Wilk test: statistic={stat:.4f}, p-value={p:.4f}")
    sns.histplot(df_11_plus_1[col])
    plt.show()
    print()
    print('От 5 до 10 лет:')
    print()
    stat, p = shapiro(df_5_to_10_1)
    print(f"Shapiro-Wilk test: statistic={stat:.4f}, p-value={p:.4f}")
    sns.histplot(df_5_to_10_1[col])
    plt.show()
    print('Меньше 4 лет:')
    print()
    stat, p = shapiro(df_4_minus_1)
    print(f"Shapiro-Wilk test: statistic={stat:.4f}, p-value={p:.4f}")
    sns.histplot(df_4_minus_1[col])
    plt.show()
```

# **Страны и регионы**

```python
states = df.groupby('state_name')['imf_pr'].agg('sum')
states_11_and_more = states[states >= 11]

print('Список стран, учавствовавших в программах МВФ более 11 лет:')
print(states_11_and_more.index.tolist())
print(f'Кол-во: {len(states_11_and_more.index.tolist())}.')
print()

states_5_to_10 = states[(states >= 5) & (states <= 10)]

print('Список стран, учавствовавших в программах МВФ от 5 до 10 лет (включительно):')
print(states_5_to_10.index.tolist())
print(f'Кол-во: {len(states_5_to_10.index.tolist())}.')
print()

states_4_and_less = states[states <= 4]

print('Список стран, учавствовавших в программах МВФ 4 и менее лет:')
print(states_4_and_less.index.tolist())
print(f'Кол-во: {len(states_4_and_less.index.tolist())}.')
```

```python
all_states = df.groupby('state_name')['region'].agg('mean')

region_sum = all_states.value_counts().sort_index()

sns.set_style('darkgrid')
ax = sns.barplot(x=region_sum.index, y=region_sum.values, color='lightgreen', edgecolor='grey')
ax.set_xticks(range(len(region_sum)))
ax.set_xticklabels(['ЛАКБ', 'С.-С. Африка', 'Ближний Восток', 'Европа', 'Пост-СССР', 'Ю.-В. Азия и АТР'], rotation=10)
plt.ylabel('count')
plt.show()
```

```python
states_11_plus = df[df.groupby('state_name')['imf_pr'].transform('sum') >= 11].groupby('state_name')['region'].agg('mean')
states_5_to_10 = df[(df.groupby('state_name')['imf_pr'].transform('sum') >= 5) & (df.groupby('state_name')['imf_pr'].transform('sum') <= 10)].groupby('state_name')['region'].agg('mean')
states_4_minus = df[df.groupby('state_name')['imf_pr'].transform('sum') <= 4].groupby('state_name')['region'].agg('mean')
```

```python
region_sum = states_4_minus.value_counts().sort_index()

sns.set_style('darkgrid')
ax = sns.barplot(x=region_sum.index, y=region_sum.values, color='lightgreen', edgecolor='grey')
ax.set_xticks(range(len(region_sum.index)))
ax.set_xticklabels(['LACB', 'S-S Africa', 'Middle East', 'Post-USSR', 'South Asia and ATR'], rotation=10)
plt.yticks([0, 4, 8, 12, 16, 20])
plt.title('Распределение стран по регионам (11+ лет в программах МВФ)')
plt.ylabel('count')
plt.show()
```

```python
region_sum = states_5_to_10.value_counts().sort_index()

sns.set_style('darkgrid')
ax = sns.barplot(x=region_sum.index, y=region_sum.values, color='lightgreen', edgecolor='grey')
ax.set_xticks(range(len(region_sum)))
ax.set_xticklabels(['LACB', 'S-S Africa', 'Middle East', 'Europe', 'Post-USSR', 'South Asia and ATR'], rotation=10)
plt.title('Распределение стран по регионам (От 5 до 10 лет в программах МВФ)')
plt.ylabel('count')
plt.show()
```

```python
region_sum = states_11_plus.value_counts().sort_index()

sns.set_style('darkgrid')
ax = sns.barplot(x=region_sum.index, y=region_sum.values, color='lightgreen', edgecolor='grey')
ax.set_xticks(range(len(region_sum.index)))
ax.set_xticklabels(['LACB', 'S-S Africa', 'Middle East', 'Europe', 'Post-USSR', 'South Asia and ATR'], rotation=10)
plt.yticks([0, 4, 8, 12, 16, 20])
plt.title('Распределение стран по регионам (11+ лет в программах МВФ)')
plt.ylabel('count')
plt.show()
```
