Формирование вспомогательного датасета с данными по жертвам политических конфликтов из трёх БД Uppsala Conflict Data Program за период 2018-2021 гг. включительно (по странам выборки).

```python
import pandas as pd

countries = ['Algeria', 'Angola', 'Benin', 'Botswana',
 'Burkina Faso', 'Burundi', 'Cabo Verde', 'Cameroon', 'Central African Republic',
 'Chad', 'Comoros', 'Congo', 'DR Congo (Zaire)', 'Ivory Coast', 'Egypt', 'Equatorial Guinea', 'Eritrea',
 'Ethiopia', 'Gabon', 'Gambia', 'Ghana', 'Guinea', 'Guinea-Bissau', 'Kenya', 'Lesotho', 'Liberia', 'Malawi',
 'Mali', 'Mauritania', 'Mauritius', 'Morocco', 'Mozambique', 'Namibia', 'Niger', 'Nigeria', 'Rwanda', 'Senegal', 
 'Sierra Leone', 'South Africa', 'Tanzania', 'Togo', 'Tunisia', 'Uganda', 'Zambia', 'Zimbabwe']
```

```python
df1 = pd.read_excel('Conflicts_databases/BattleDeaths_v24_1.xlsx')
df2 = pd.read_excel('Conflicts_databases/Nonstate_v24_1.xlsx')
df3 = pd.read_excel('Conflicts_databases/OneSided_v24_1.xlsx')
```

```python
df1 = df1[['location_inc', 'year', 'bd_best']].rename(columns={'location_inc': 'location', 'bd_best': 'best_fatality_estimate'})
df2 = df2[['location', 'year', 'best_fatality_estimate']]
df3 = df3[['location', 'year', 'best_fatality_estimate']]
```

```python
df1 = df1[df1['year'].isin([2019, 2020, 2021]) & df1['location'].isin(countries)]
df2 = df2[df2['year'].isin([2019, 2020, 2021]) & df2['location'].isin(countries)]
df3 = df3[df3['year'].isin([2019, 2020, 2021]) & df3['location'].isin(countries)]
```

```python
grouped_df1 = df1.groupby(['location', 'year']).agg(fatalities_estimate=('best_fatality_estimate', 'sum'))
grouped_df2 = df2.groupby(['location', 'year']).agg(fatalities_estimate=('best_fatality_estimate', 'sum'))
grouped_df3 = df3.groupby(['location', 'year']).agg(fatalities_estimate=('best_fatality_estimate', 'sum'))

merged_df = (grouped_df1.merge(grouped_df2, how='outer', on=['location', 'year'])
             .merge(grouped_df3, how='outer', on=['location', 'year']).fillna(0)
             .rename(columns={'fatalities_estimate_x': 'fatalities_estimate_2020',
                              'fatalities_estimate_y': 'fatalities_estimate_2021',
                              'fatalities_estimate': 'fatalities_estimate_2022'}))

merged_df['overall_fatalities'] = merged_df['fatalities_estimate_2020'] + merged_df['fatalities_estimate_2021'] + merged_df['fatalities_estimate_2022']
merged_df
```
